```python
import pandas as pd

df = pd.read_csv("../data/mbti_data_まちこ_INFP_まちこ_INFP_all_data_20250623_130040.csv")
```

```python
df.head()
```

```python
requests = []
element_map = {
    "エネルギー (I/E)": 0,
    "情報収集 (N/S)": 1,
    "意思決定 (T/F)": 2,
    "外界への態度 (J/P)": 3,
}
for phase, d in df.groupby("Phase"):
    element_id = element_map[d.iloc[0]["Element Type"]]
    records = (
        d[["Question", "Answer"]]
        .rename({"Question": "assistant", "Answer": "user"}, axis=1)
        .to_dict(orient="records")
    )
    # Convert each record to [{"role": "assistant", "content": ...}, {"role": "user", "content": ...}]
    messages = []
    for r in records:
        messages.append({"role": "assistant", "content": r["assistant"] if pd.notna(r["assistant"]) else "nan"})
        messages.append(
            {
                "role": "user",
                "content": r["user"] if pd.notna(r["assistant"]) else "nan",
            }
        )
    requests.append(
        {
            "element_id": element_id,
            "messages": messages,
        }
    )
```

```python
import json
# parse japanese

print(json.dumps(requests, ensure_ascii=False, indent=2))
```
